```python
def toCommon(district):
    if district.endswith('Board Of Education School District'):
        district = district[:-35]
    elif district.endswith('Public School District'):
        district = district[:-23]
    elif district.endswith('Public Schools'):
        district = district[:-15]
    elif district.endswith('School District'):
        district = district[:-16]
    return district
```

```python
districts_abbr = [toCommon(d) for d in districts]
```

```python
df_niche['mark2'] = df_niche.district.apply(lambda d: toCommon(d) in districts_abbr)
```

```python
df_niche['common'] = df_niche.district.apply(toCommon)
df_gs['common'] = df_gs.district.apply(toCommon)
```

```python
def preprocess(df, extra_col):
    df_by_district_rating = df.groupby(['common',extra_col]).school.count().to_frame()
    df_by_district = df.groupby('common').school.count().to_frame()
    
    df_by_district_rating = df_by_district_rating.unstack(level=1)
    df_by_district_rating.columns = [col[-1] for col in df_by_district_rating.columns]
    df_by_district = df_by_district_rating.merge(df_by_district, left_index=True, right_index=True)
    df_by_district = df_by_district.fillna(0).astype(int)
    
    return df_by_district
```

```python
df_counts_niche = preprocess(df_niche, 'rating')
df_counts_gs = preprocess(df_gs, 'rating_scores')
```

```python
df_counts_niche = df_counts_niche.reset_index()
df_counts_gs = df_counts_gs.reset_index()
```

```python
df_counts_niche['mark'] = df_counts_niche.common.apply(lambda s: s in districts_abbr)
df_counts_gs['mark'] = df_counts_gs.common.apply(lambda s: s in districts_abbr)
```
